```python
import time
import numpy as np
import matplotlib.pyplot as plt
```

```python
def func(n):
    if n <= 1:
        return n
    return func(n-1) + func(n-2)
```

```python
r_start_time = time.time() 
r_num_list = []
r_exec_time = []
for n in range(0,25):
    r_num_list.append(func(n))
    r_exec_time.append(time.time() - r_start_time)
print(f'recursion fibonacci series: {r_num_list}')
print(f'recursion execution time: {r_exec_time}')
```

```python
plt.plot(r_num_list, r_exec_time)
plt.xlabel('Fibonacci series')
plt.ylabel('Time Taken to Generate')
plt.title('Fibonacci series with Recursion')
plt.show()
```

```python
first_num = 0
second_num = 1
count = 0
n = 25
start_time = time.time()
num_list =[]
exec_time = []
while count< n:
    # print(first_num)
    num_list.append(first_num)
    exec_time.append(time.time() - start_time)
    first_num, second_num = second_num, first_num + second_num
    count += 1
    
print(f'fibonacci series: {num_list}')
print(f'execution time: {exec_time}')        
```

```python
plt.plot(num_list, exec_time)
plt.xlabel('Fibonacci series')
plt.ylabel('Time Taken to Generate')
plt.title('Fibonacci series without Recursion')
plt.show()
```

```python
plt.plot(r_num_list, r_exec_time, label ='Recursion')
plt.plot(num_list, exec_time, label = "No-Recursion")
plt.xlabel('Fibonacci series')
plt.ylabel('Time Taken to Generate')
plt.title('Comparison')
plt.legend()
plt.show()
```

Difference between the recursion and non resursion(iteration) can be seen when we observe the time taken to generte the fibonacci series as recursion takes up more time to generate the series as we iterate through each recursion whereas non-recursion or iterative method is considerably faster. Also upon further research, I came to know that recursion takes considerably more space/memory than compared to iteration when generating a larger series as the space required is proportional to the max depth of the recursion tree.
